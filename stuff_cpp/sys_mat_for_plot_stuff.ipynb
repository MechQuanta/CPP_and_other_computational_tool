```python
import pandas as pd
import matplotlib.pyplot as plt

# Read the CSV file into a pandas DataFrame
df = pd.read_csv("system_data.csv", header=None, names=["Index", "t", "x", "y"])

# Create the plot
plt.figure(figsize=(10, 6))  # Adjust figure size if needed
plt.plot(df["x"], df["y"])

# Customize the plot
plt.xlabel("x")
plt.ylabel("y")
plt.title("Plot of x vs y")
plt.grid(True)

# Show the plot
plt.show()
```

```python
import csv

with open('damping_diffeq.dat', 'r') as dat_file:
    with open('system_data2.csv', 'w', newline='') as csv_file:
        csv_writer = csv.writer(csv_file)
        for row in dat_file:
            row = [value.strip() for value in row.split('|')]
            csv_writer.writerow(row)
```
